```python
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(-10, 10)
y = 2*x + 1

y1 = 6*x - 2
y2 = 0.1*x+6

plt.figure()
plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim(-2, 10)
plt.ylim(-2, 10)
# draw axes
plt.axvline(x=0, color='#A9A9A9')
plt.axhline(y=0, color='#A9A9A9')
plt.show()
plt.close()
```

#### Pseudoinverse

 $A^+ = (A^TA)^{-1}A^T$

```python
A = np.array([[-2,1],[-6,1],[-1/10,1]])
```

```python
Ap = A.T@A
Ap
```

```python
np.linalg.matrix_rank(Ap)
```

```python
x = np.linalg.inv(Ap)@A.T@np.array([[1],[-2],[6]])
x
```
